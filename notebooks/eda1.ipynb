```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load your data (replace with your file paths)
df = pd.read_csv("../data/manual/allstar_data.csv")
# Example plot - number of All-Star players by year
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x="year")
plt.title("Number of All-Star Players by Year")
plt.show()
```

```python
# Basic statistics
df.describe()

# Data info to check for missing values and data types
df.info()
```

```python
# Check for missing data
df.isnull().sum()

# Drop rows with missing values (if appropriate)
df_cleaned = df.dropna()

```

```python
# All-Star selections by team
team_counts = df['team'].value_counts()

# Plot
plt.figure(figsize=(12, 6))
sns.barplot(x=team_counts.index, y=team_counts.values)
plt.xticks(rotation=90)
plt.title("All-Star Selections by Team")
plt.show()
```
